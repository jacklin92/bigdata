```python
from collections import Counter
from typing import List, Tuple

def ne_word_frequency(a_news_ne: List[Tuple[str, str]], top_k: int = 200) -> List[Tuple[str, int]]:
    """
    計算命名實體出現次數，僅限長度大於等於2且類型在 allowedNE 中的詞

    :param a_news_ne: List of tuples like (NER tag, word)
    :param top_k: Number of top frequent words to return
    :return: List of (word, frequency) tuples
    """
    filtered_words = [
        word for ner, word in a_news_ne
        if len(word) >= 2 and ner in allowedNE
    ]
    counter = Counter(filtered_words)
    return counter.most_common(top_k)
```

```python
# All-in-one function
def get_top_ner_words():
    
    top_cate_ner_words={} # final result
    counter_all = Counter() # counter for category '全部'
    for category in news_categories:

        df_group = df[df.category == category]

        # concatenate all filtered words in the same category
        words_group = []
        for row in df_group.entities:

            # filter words for each news
            filtered_words =[]
            for (ner, word) in eval(row):
                if (len(word) >= 2) & (ner in allowedNE):
                    filtered_words.append(word)

            # concatenate filtered words  
            words_group += filtered_words

        # now we can count word frequency
        counter = Counter( words_group )

        # combine counter_all with conter
        counter_all += counter
        
        # get a few most common items
        topwords = counter.most_common(200)

        # store topwords
        top_cate_ner_words[category]= topwords

    # Process category '全部'
    top_cate_ner_words['全部'] = counter_all.most_common(200)
    return list(top_cate_ner_words.items())
```

```python
from collections import defaultdict, Counter

def group_ne_by_type_and_category(df):
    # 建立一個 dict 結構：{ ner_type: { category: Counter(...) } }
    ner_type_dict = defaultdict(lambda: defaultdict(Counter))

    for _, row in df.iterrows():
        category = row['category']
        entities = eval(row['entities'])  # 這裡是 list of (ner, word)

        for ner, word in entities:
            if len(word) >= 2 and ner in allowedNE:
                ner_type_dict[ner][category][word] += 1
                ner_type_dict[ner]['全部'][word] += 1  # 統計總和

    # 把資料轉成你要的格式
    result = []
    for ner_type in sorted(ner_type_dict.keys()):
        cate_list = []
        for category in news_categories + ['全部']:
            top_items = ner_type_dict[ner_type][category].most_common(200)
            cate_list.append((category, top_items))
        result.append((ner_type, cate_list))

    return result
```

```python
def ner_result_to_df(ner_result):
    rows = []
    for ner_type, cate_list in ner_result:
        rows.append({
            'ne_name': ner_type,
            'top_keys': str(cate_list)  # 轉成字串儲存
        })
    return pd.DataFrame(rows)
```
